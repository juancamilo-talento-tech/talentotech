**REGRESION LINEAL**

```python
import pandas as pd

# Definir los valores para las variables x e y
valores_x = [1, 2, 2, 3, 4, 4, 5, 6]
valores_y = [2, 3, 4, 4, 4, 6, 5, 7]

# Crear un DataFrame con Pandas
df = pd.DataFrame({'x': valores_x, 'y': valores_y})

df['xy'] = df['x'] * df['y']
df['xx']= df['x']**2

# Calcular la sumatorias
suma_x = df['x'].sum()
suma_y = df['y'].sum()
suma_xy = df['xy'].sum()
suma_x2 = df['xx'].sum()

# Imprimir los resultados
print("Suma de x:", suma_x)
print("Suma de y:", suma_y)
print("Suma de xy:", suma_xy)
print("Suma de xx:", suma_x2)



# Añadir una fila con las sumatorias al DataFrame
df.loc['Sumatoria'] = pd.Series({'x': suma_x, 'y': suma_y, 'xy': suma_xy, 'xx': suma_x2})

# Mostrar el DataFrame con la nueva fila
df
```

```python
#Para calcular la pendiente
n= df.shape[0]-1
print(n)
m= (n*suma_xy-(suma_x)*(suma_y))/(n*suma_x2-(suma_x)**2)

# Calcular el promedio de y
promedio_x = df['x'].mean()
promedio_y = df['y'].mean()
print("Promedio de x:", promedio_x)
print("Promedio de y:", promedio_y)
#promedio_x
#promedio
b= (suma_y)/n-(m*suma_x/n)
b
```

```python
import matplotlib.pyplot as plt
# Graficar x vs y
df1 = df.iloc[:-1]  # Eliminar la última fila (Sumatoria)
# print(df1)        #imprime el dataframe 1 sin la ultima fila
y=m*df1['x']+b
plt.figure(figsize=(8, 6))  # Tamaño de la figura
plt.scatter(df1['x'], df1['y'], color='blue', label='Datos')
plt.plot(df1['x'], y, color='red', label='Regresión Lineal')

plt.title('Regresion Lineal')  # Título del gráfico
plt.xlabel('x')  # Etiqueta del eje x
plt.ylabel('y')  # Etiqueta del eje y
plt.grid(True)  # Mostrar cuadrícula
plt.legend()  # Mostrar leyenda
plt.show()  # Mostrar gráfico
```

```python
import numpy as np
# Caulcular ypred
df['ypred'] = m*df1['x']+b

# calcular error absoluto

df['error_abs'] = np.abs(df['y'] - df['ypred'])

# calcular error cuadratico
df['error_cuad'] = (df['y'] - df['ypred'])**2
print(df)

sr=df['error_cuad'].sum()
print('El valor de Sr=',sr)

```
